# Filtering exercises with Stackoverflow 2022 dataset

```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df11 = pd.DataFrame(people)
df11
```

```python
# People whose last name is Doe
df11[df11["last"]== "Doe"]
```

```python
df11.loc[df11["last"]=="Doe"]
```

```python
# emails of People whose last name is Doe

df11[df11["last"]== "Doe"]["email"]
```

```python
df11.loc[df11["last"]=="Doe", "email"]
```

```python
# people with last name of Doe and first name of John

df11.loc[(df11["last"]=="Doe")&(df11["first"]=="John")]
```

```python
# email of people with last name of Doe and first name of John
df11.loc[((df11["last"] == "Doe") & (df11["first"] == "John")), "email"]
```

```python
# last name is equal to Schafer or first name is equal to john
df11.loc[(df11["last"] == "Schafer") | (df11["first"] == "John")]
```

```python
# email of last name is equal to Schafer or first name is equal to john
df11.loc[((df11["last"] == "Schafer") | (df11["first"] == "John")), "email"]
```

```python
# if we want to bring all the emails of people whose last name is not schafer and f-name is not john,
# 1) we can use != for entire code
# 2) we can basicly use a tilde ~

df11.loc[~((df11["last"] == "Schafer") | (df11["first"] == "John")), "email"]
```

# Splitting object >> applying function >> combining results : Groupby

```python
# let's groups the data throguh countries. 
# 1. split the object: let's check the countries in the data

df["Country"].value_counts()
```

```python
df.groupby(["Country"])
```

```python
country_grp = df.groupby(["Country"])
```

```python
# now we can get some specific sub groups from this country group. let's check Germany
country_grp.get_group("Germany")

# this is same with df["Country"]== "Germany"
```

```python
# 2 - apply the function

# we've already checked the median salary for the entire data above. However, purchasing power and living 
# costs differ across countries significantly. So we can't have efficient 
# impressions from median. So let's group the average salaries by countries. In other words: apply mean function

df.groupby(["Country"])["ConvertedCompYearly"].mean()
```

```python
# let's group countries for each age group
country_grp["Age"].value_counts()
```

```python
# let's check some specific countries
country_grp["Age"].value_counts().loc["Germany"]
```

```python
country_grp["Age"].value_counts(normalize=True).loc["Germany"]
```

```python
country_grp["Age"].value_counts(normalize=False).loc["United Kingdom of Great Britain and Northern Ireland"]
```

```python
# this time let's age groups by countries
df.groupby("Age")["Country"].value_counts()
```

```python
# let's check age column to see why can't we see the median of age which means splitting Age group
df["Age"].value_counts()
```

```python
# now we can check average salary for each age group
df.groupby(["Age"])["ConvertedCompYearly"].mean()
```

```python
df.groupby("Country")["ConvertedCompYearly"].median()

# bu artık indeksleri ülke olan bir veri. dolayısıyla en soan indeks olarak ülke yazarak ülkelere bakabiliriz
```

```python
# what percentage of people from each country use python?

# first we need the number of respondants from each country for that

print(df["Country"].value_counts())
# let's assign it to a variable

country_filt = df["Country"].value_counts()
```

```python
# second: let's find python users for each country
print(df.groupby("Country")["LanguageHaveWorkedWith"].apply(lambda x: x.str.contains("Python").sum()))

country_pyt_users = df.groupby("Country")["LanguageHaveWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
```

```python
# 3. 
# 3.1 traditional way: but it's too difficult to read and analysis
np.ceil((country_pyt_users / country_filt)*100)
```

```python
# 3.2 concat method and creating a new dataframa

python_df = pd.concat([country_filt, country_pyt_users], axis = "columns", sort = False)
python_df
```

```python
# now let's change the column names

python_df.rename(columns = {"Country": "NumRespondants", "LanguageHaveWorkedWith": "NumKnowsPython"}, inplace=True)
python_df
```

```python
# last, let's add a column for percentages
python_df["Percentage_Pyt_Users"] = (python_df["NumKnowsPython"] / python_df["NumRespondants"])*100
python_df
```

```python
# now we have a new df and we can do all operations that we can do for a df. let's sort it through respondants

python_df.sort_values("NumRespondants")
```
